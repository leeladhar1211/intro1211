## Treating Missing Values

```python
import numpy as np
np.nan
```

```python
np.nan == np.nan
```

```python
data.head()
```

```python
data.info() # column type give hint
```

```python
data.qsec.sort_values(ascending=True)
```

```python
data.head()
```

```python
data.replace(['?','*','-','$'],np.nan,inplace=True)
```

```python
data.fillna(method='bfill') 
# bfill , ffill is usually used for categorical variables
```

```python
data.info()
```

```python
data.disp = data.disp.astype(float)
data.wt = data.wt.astype(float)
data.qsec = data.qsec.astype(float)
data.info()
```

```python
data.plot(kind='box')
```

```python
%matplotlib inline
data.mpg.plot(kind='box')
```

```python
data.loc[data.mpg>100,'mpg'] = np.nan
```

```python
data.mpg.plot(kind='box')
```

```python
data.head()
```

```python
data.fillna(data.mean(),inplace=True)
```
